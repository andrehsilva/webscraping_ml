```python
import pandas as pd
import sqlite3
from datetime import datetime
```

```python
df_note = pd.read_json('../output/data.json')
df_note.head()
```

```python
df_smart = pd.read_json('../output/data_smartphone.json')
df_smart.head()
```

```python
def rename_columns(df):
    df = df.rename(columns={
        'brand': 'Marca',
        'name': 'Descrição',
        'seller': 'Vendido por',
        'reviews_rating_number': 'Avaliações',
        'reviews_amount': 'Quantidade de Avaliações',
        'old_money': 'Preço Antigo',
        'new_money': 'Preço Novo',
    })
    
    
    # Transforma nomes das colunas em maiúsculas
    df.columns = [col.upper() for col in df.columns]
    return df
```

```python
df_note = rename_columns(df_note)
df_note.head()
```

```python
df_smart = rename_columns(df_smart)
df_smart.head()
```

```python
def ajuste_columns(df):
    # Remove espaços em branco do início e do fim das strings
    df = df.map(lambda x: x.strip() if isinstance(x, str) else x)
    
    # Remove os parênteses da coluna (mas mantém o conteúdo)
    df['QUANTIDADE DE AVALIAÇÕES'] = df['QUANTIDADE DE AVALIAÇÕES'].replace(r'[()]', '', regex=True)
    
    # Converte colunas para tipos numéricos
    df['PREÇO ANTIGO'] = df['PREÇO ANTIGO'].astype(float)
    df['PREÇO NOVO'] = df['PREÇO NOVO'].astype(float)

    return df
```

```python
df_note.describe()
```

```python
df_note = ajuste_columns(df_note)
df_note.head()
```

```python
df_note.dtypes
```

```python
df_smart = ajuste_columns(df_smart)

df_smart.head()
```

```python
def substituir_marca(df, palavras):
    for palavra in palavras:
        # Substitui None em MARCA se a palavra estiver presente na DESCRIÇÃO
        df.loc[(df['MARCA'].isna()) & (df['DESCRIÇÃO'].str.contains(palavra, case=False, na=False)), 'MARCA'] = palavra

    # Preenche valores restantes como "Indefinido"
    df['MARCA'] = df['MARCA'].fillna('Indefinido')
    return df

# Lista de palavras para verificar
palavras = ['Xiaomi', 'Poco','LG','Samsung', 'Infinix','Honor','Motorola', 'Realme', 'Zte','Multilaser', 'Apple', 'Asus', 'Lenovo', 'LG', 'Nokia', 'Sony', 'Philips', 'Positivo', 'Acer', 'Dell', 'HP', 'Vaio', 'Microsoft', 'Samsung', 'Multilaser']

```

```python
# Aplicar a função no dataframe df_smart
df_smart = substituir_marca(df_smart, palavras)
df_smart.head()
```

```python
df_note = substituir_marca(df_note, palavras)
df_note.head()
```

```python
df_note['TIPO'] = 'Notebook'
df_smart['TIPO'] = 'Smartphone'
```

```python
df_smart
```

```python
df = pd.concat([df_note, df_smart], ignore_index=True)
df.head()
```

```python
df.to_csv('../output/df_final.csv', index=False)
df.to_json('../output/df_final.json', orient='records', lines=True)
```

```python
# 1. Garante que colunas estejam numéricas
df['QUANTIDADE DE AVALIAÇÕES'] = pd.to_numeric(df['QUANTIDADE DE AVALIAÇÕES'], errors='coerce')
df['AVALIAÇÕES'] = pd.to_numeric(df['AVALIAÇÕES'], errors='coerce')
df['PREÇO ANTIGO'] = pd.to_numeric(df['PREÇO ANTIGO'], errors='coerce')
df['PREÇO NOVO'] = pd.to_numeric(df['PREÇO NOVO'], errors='coerce')

# 2. Agrupa os dados por marca e tipo
df_grouped = df.groupby(['MARCA', 'TIPO']).agg(
    QTD_PRODUTOS=('DESCRIÇÃO', 'count'),  # Conta quantos produtos tem por grupo
    MEDIA_PRECO_ANTIGO=('PREÇO ANTIGO', 'mean'),
    MEDIA_PRECO_NOVO=('PREÇO NOVO', 'mean'),
    SOMA_AVALIACOES=('QUANTIDADE DE AVALIAÇÕES', 'sum'),
    MEDIA_AVALIACOES=('AVALIAÇÕES', 'mean')
).reset_index()

# 3. Formata os dados
df_grouped['MEDIA_PRECO_ANTIGO'] = df_grouped['MEDIA_PRECO_ANTIGO'].round(2)
df_grouped['MEDIA_PRECO_NOVO'] = df_grouped['MEDIA_PRECO_NOVO'].round(2)
df_grouped['MEDIA_AVALIACOES'] = df_grouped['MEDIA_AVALIACOES'].round(2)
df_grouped['SOMA_AVALIACOES'] = df_grouped['SOMA_AVALIACOES'].fillna(0).astype(int)

# 4. Exibe o relatório
df_grouped.head()
```
